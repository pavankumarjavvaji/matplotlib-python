```python
from collections import Counter
import pandas as pd
data=pd.read_csv('data.csv')
data.head()
```

```python
ids=data['Responder_id']
lan_res=data['LanguagesWorkedWith']
language_counter=Counter()
for res in lan_res:
    language_counter.update(res.split(';'))
language_counter
```

```python
lang=[]
cou=[]
for a in language_counter.most_common():
    lang.append(a[0])
    cou.append(a[1])
lang.reverse()
cou.reverse()
```

```python
plt.barh(lang,cou)
```

```python
from itertools import count
from matplotlib.animation import FuncAnimation
import random
x_vals=[]
y_vals=[]
index=count()
def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0,5))
    plt.cla()
    plt.plot(x_vals,y_vals)
ani=FuncAnimation(plt.gcf(),animate,interval=1000)
```

```python
import csv
import time
x_value=0
total_1=1000
total_2=1000
fieldnames=["x_value","total_1","total_2"]
with open('realdata.csv','w') as csv_file:
    csv_writer=csv.DictWriter(csv_file,fieldnames=fieldnames)
    csv_writer.writeheader()
while True:
    with open('realdata.csv','a') as csv_file:
        csv_writer=csv.DictWriter(csv_file,fieldnames=fieldnames)
        info={
            "x_value":x_value,
            "total_1":total_1,
            "total_2":total_2
        }
        csv_writer.writerow(info)
        x_value+=1
        total_1+=random.randint(-6,8)
        total_2+=random.randint(-5,6)
    time.sleep(1)
```
